```python
import pandas as pd
import folium as fl
```

```python
# Import streamflow data
data = pd.read_csv('../data/streamflow_q20_complete.csv', dtype={'STATION_NO':str, 'HUC8_1':str})
data = data[0:500]
```

```python
# Draw gage data in map
def plot_map_year(year):
    # Open map
    fmap = fl.Map(location=[38.5, -95], zoom_start=4, width='70%', height='70%')
    for i,row in data.iterrows():
        popup_text = '{}<br> Station number: {}<br> Drainage area: {}<br> HUC8: {}<br> URL: {}<br> Record count: {}<br> Flow (20th perc): {}'
        popup_text = popup_text.format(row['STATION_NM'], row['STATION_NO'], row['DRAIN_AREA'], 
                                       row['HUC8_1']+' - '+row['NAME'], row['NWIS_URL'], 
                                       row['Count_'+str(year)], row['Flow_p20_'+str(year)])
        #fl.CircleMarker(location=(row['LATDD'], row['LONDD']), radius=1, color='#007849', popup=popup_text
        #                , fill=True).add_to(fmap)
        fl.Marker(location=(row['LATDD'], row['LONDD']), popup=popup_text).add_to(fmap)
    return fmap
```

```python
plot_map_year(2012)
```

```python
fmap.save('test.html')
```
